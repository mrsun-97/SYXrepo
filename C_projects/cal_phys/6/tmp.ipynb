由球盖立体角公式
$$\Omega~=~2\pi\frac{h}{R}$$
知球面随机点在$x$轴上投影是均匀分布的，

```julia
using PyPlot
using DelimitedFiles
arr = readdlm("data.txt", '\t')
len = size(arr, 1)
X = arr[:,1]
Y = arr[:,2]
Z = arr[:,3]
c = Int(2)
println("len = ",len)
```

```julia
fig=figure(figsize=[12, 12])
grid("on")
xlabel("x")
ylabel("y")
zlabel("z")
scatter3D(0,0,0,s=20,c="red")
scatter3D(X,Y,Z,s=1,c="blue")
name = "sphere"
savefig("./$name$c-3d.png")
f = open("./$name$c.dat", "w")
for i = 1:len
    writedlm(f,[X[i] Y[i] Z[i]])
end
close(f)

si = 5
fig2=figure(figsize=[si, si])
grid("on")
xlabel("x")
ylabel("z")
scatter(0,0,s=20,c="red")
scatter(X,Z,s=1,c="blue")
savefig("./$name$c-xz.png")

fig3=figure(figsize=[si, si])
grid("on")
xlabel("x")
ylabel("y")
scatter(0,0,s=20,c="red")
scatter(X,Y,s=1,c="blue")
savefig("./$name$c-xy.png")
```
