```julia
using PyPlot
using DelimitedFiles

arr = readdlm("data.txt", '\t')
Len = arr[1,1]
Hei = arr[1,2]
arr = arr[2:end,:]
art = arr'
println(Len,"\t",Hei,"\tDone!")
```

```julia
fig = figure(figsize=[10, 10])
grid("on")
xlabel("x")
xlim(200,800)
#xlim(0,Len)
#xticks(0:Len/5:Len)
ylabel("y")
ylim(200,800)
#ylim(0,Hei)
#yticks(0:Hei/5:Hei)
scatter(art[1,:], art[2,:],s=2)
name = "[336]"
c = "1"
savefig("./$name$c.jpg")
f = open("./$name$c.dat", "a")
for i = 1:size(arr,1)
    writedlm(f,[arr[i,1] arr[i,2]])
end
a = b = 0
close(f)
```
