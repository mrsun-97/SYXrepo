<h2>sandbox法计算分形维数</h2>

```julia
using PyPlot
using DelimitedFiles
k = readdlm("sandbox.txt",'\t')
r = k'[1,2:148]
rs= log.(k'[1,:])
N = k'[2,2:148]
Ns= log.(k'[2,:])

n = length(r)
r = log.(r)
N = log.(N)
r̄ = sum(r)/n
N̄ = sum(N)/n
a = (r'*N-n*r̄*N̄)/(r'*r-n*r̄^2)
b = N̄ - a*r̄
R_square = (r'*N-n*r̄*N̄)^2/((r'*r-n*r̄^2)*(N'*N-n*N̄^2))
println("y = ax + b:\n","a = ",a,"\nb = ",b,"\nR^2 =",R_square)
```

```julia
fig2 = figure(figsize=[10, 10])
grid("on")
title("sandbox\ndim = $a")
xlabel("ln(r)")
ylabel("ln(N)")
scatter(rs,Ns,s=10)
x = 0.2:5.2
plot(x,a*x.+b,c="red")
savefig("./[336]1sandbox.png")
```
