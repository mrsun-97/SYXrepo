```julia
using PyPlot, DelimitedFiles
bigarr = readdlm("data.txt",'\t')
M = size(bigarr,1)
N = Int(size(bigarr,2)/3)
```

```julia
C = zeros(N)
sumn = 0;
for i in 1:N
    sumn = 0
    for j in 1:M
        sumn += cos(bigarr[j,3i-1])*cos(bigarr[j,3i-1])
    end
    C[i] = sumn/M
end
figC = figure(figsize=[10,10])
grid()
plot(1:length(C),C)
show()
```
