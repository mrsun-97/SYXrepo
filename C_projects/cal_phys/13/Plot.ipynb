```julia
using PyPlot, DelimitedFiles
bigarr = readdlm("data.txt",'\t')
M = size(bigarr,1)
N = Int(size(bigarr,2)/3)
```

```julia
C = zeros(N)
sumn = 0;
for i in 1:N
    sumn = 0
    for j in 1:M
        sumn += cos(bigarr[j,2])*cos(bigarr[j,3i-1])
    end
    C[i] = sumn/M
end
figC = figure(figsize=[10,10])
grid()
xlabel("steps")
ylabel("C(l)")
plot(0:20:20length(C)-1,C)
show()
c = "dot3"
savefig("./texfile/exp$c.png")
figlog = figure(figsize=[10,10])
grid()
xlabel("steps")
ylabel("log10[C(l)]")
plot(0:20:20length(C)-1,log10.(C))
savefig("./texfile/log$c.png")
```
