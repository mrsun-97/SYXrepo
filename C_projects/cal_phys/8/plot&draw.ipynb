```julia
using PyPlot
using DelimitedFiles
arr = readdlm("data0.txt", '\t')
A = readdlm("lattice.txt", '\t')
len = size(arr, 1)

c = "5"

arr[1,1]=0
x = arr[:,1]
y = arr[:,2]
ed = " "
if y[1] >=50
    ym = minimum(y)
    ed = "min"
else
    ym = maximum(y)
    ed = "max"
end
println("len = ",len)
println("c = ",c)
```

```julia
fig=figure(figsize=[12, 8])
grid("on")

xlabel("t")
ylabel("%")
xlim(-2,len*1-4)
ylim(1,8)
yticks(1:8)
title("|↑> start at $(y[1])%, $ed:$ym%")
plot(x,y)
plot(x,ym*ones(len),c="red")
plot(1:Int(ceil(len/2)),y[1]*ones(Int(ceil(len/2))))
name = "data"
savefig("./texfile/$name$c.png")
f = open("./texfile/$name$c.dat", "w")
for i = 1:len
    writedlm(f,[x[i] y[i]])
end
close(f)
```
