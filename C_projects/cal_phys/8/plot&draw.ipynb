```julia
using PyPlot
using DelimitedFiles
arr = readdlm("data0.txt", '\t')
len = size(arr, 1)
c = Int(1)
arr[1,1]=0
x = arr[:,1]
y = arr[:,2]
ym =  maximum(y)
println("len = ",len)
```

```julia
fig=figure(figsize=[12, 8])
grid("on")
xlabel("t")
ylabel("%")
xlim(0,len*5-4)
ylim(0,100,10)
yticks(0:10:100)
plot(x,y)
plot(x,ym*ones(len),c="red")
plot(1:len,y[1]*ones(len))
name = "data"
savefig("./$name$c.png")
f = open("./$name$c.dat", "w")
for i = 1:len
    writedlm(f,[x[i] y[i]])
end
close(f)
```
