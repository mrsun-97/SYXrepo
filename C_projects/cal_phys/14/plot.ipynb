```julia
using PyPlot, DelimitedFiles
c = 3
e = [1.0,1.0,0.3425]
arr = readdlm("data_$c-1.txt",'\t')
x = arr[:,1]
y = -arr[:,2].+e[c]
println("finished")
```

```julia
r = x
N = y
n = length(r)
r = log10.(r)
N = log10.(N)
r̄ = sum(r)/n
N̄ = sum(N)/n
a = (r'*N-n*r̄*N̄)/(r'*r-n*r̄^2)
b = N̄ - a*r̄
R_square = (r'*N-n*r̄*N̄)^2/((r'*r-n*r̄^2)*(N'*N-n*N̄^2))
println("y = ax + b:\n","a = ",a,"\nb = ",b,"\nR^2 = ",R_square)
```

```julia
fig = figure(figsize=[10,10])
grid()
xlabel("log10(N)")
ylabel("log10(0.3425-Pd(N))")
title("dim=$c, k=$a")
plot(log10.(x), log10.(y))
plot(log10.(x),a*log10.(x).+b)
savefig("./texfile/log_$c.png")
```
